#### Model kullanarak secim yapma
Simdiye kadar gordugumuz metotlar, değişkenleri sıralamak ve ardından uyum iyiliğine göre özellikleri seçmek için istatistiksel testlerin yöntem olarak kullanımını tanıttık. Bu bölümde, değişkenleri seçmek için değişken sıralamasından yararlanan filtre yöntemlerini tartışmaya devam edeceğiz. Model performansının bir metriğine dayalı olarak özellikleri niteleyen sıralama yöntemlerini tartışacağız. 
Her ozellik uzerine bir model kurup performans metrigini o degisken icin bir skor olarak gorebiliriz. 

```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from feature_engine.selection import SelectBySingleFeaturePerformance
from sklearn.metrics import roc_auc_score, mean_squared_error
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
x = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
y = breast_cancer.target

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)

selector= SelectBySingleFeaturePerformance(estimator=DecisionTreeClassifier(random_state=10), scoring="roc_auc",cv=3,threshold=0.8,)

selector.fit(X_train,y_train)
selector.feature_performance_

pd.Series(selector.feature_performance_).sort_values(ascending=False).plot.bar(figsize=(10, 5))
plt.xlabel('degiskenler')
plt.ylabel('ROC AUC')
plt.title('The bigger the better');
```

```python
X_train_t=selector.transform(X_train)
X_test_t=selector.transform(X_test)
```

```python
X_train_t.columns
```

#### Wrapper Methods- Forward and Backward
Regresyon veya siniflandirma modelleri uzerine sarilmis bir metotdur. Bu metot icinde, ayni modeli butun ozelliklerin farkli kominasyonlarindan olusan alt ozellik kumesine uygulanarak elde edilen performans metriklerinin sonucuna gore bir skorlama islemi yapilir. Bu yontemin basit olmasi avantaj iken en buyuk dezavantaji hesaplama zamanidir. Ornegin, $n$ tane degiskeni olan bir veride $2^n$ taneden olusan alt kumelerde secim yapilarak model olusturulur. 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SequentialFeatureSelector as SFS
from sklearn.metrics import roc_auc_score
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
x = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
y = breast_cancer.target

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)

selector=SFS(estimator=RandomForestClassifier(n_estimators=3, random_state=0),
             n_features_to_select='auto',
             scoring='roc_auc',
             cv=2,
    direction='forward',)
selector.fit(X_train,y_train)
```

```python
X_train.columns
```

```python
selector.get_feature_names_out()
```

```python
X_train_t=selector.transform(X_train)
X_test_t=selector.transform(X_test)
X_train_t.head()
```

#### Gömülü Yöntemler (Embedded Methods)
Gömülü Yöntemler (Embedded Methods), özellik seçimi (feature selection) sürecinin bir parçası olarak, bir öğrenme algoritmasının eğitilmesi sırasında özelliklerin önemini belirlemek için kullanılan yöntemlerdir. Bu yöntemler, özelliklerin model performansına doğrudan katkı sağlayıp sağlamadığını belirlemek için kullanılır.

Gömülü yöntemler, filtre yöntemleri ve sarınma (wrapper) yöntemleri arasında bir ara noktada yer alır. Bunlar, filtre yöntemlerinden farklı olarak, özelliklerin önemini belirlemek için ayrı bir özellik seçimi adımı gerektirmezler. Bunun yerine, eğitim sırasında modelin içsel olarak özellikleri seçmesine izin verirler.

Bir öğrenme algoritması eğitilirken, gömülü yöntemler genellikle aşağıdaki şekillerde kullanılır:

1. **Özellik Ağırlıkları (Feature Weights):**
   Bazı algoritmalar, özelliklerin önemini belirlemek için her bir özellik için bir ağırlık veya önem derecesi atar. Örneğin, lineer regresyon veya destek vektör makineleri gibi algoritmalar, her bir özellik için bir katsayı (ağırlık) belirler. Bu katsayılar, özelliklerin modelin çıktısına olan katkısını gösterir.

2. **Özellik Önemi Ölçüleri (Feature Importance Measures):**
   Bazı ağaç tabanlı yöntemler, her bir özelliğin ne kadar sıklıkla kullanıldığını veya bir düğümde ne kadar önemli olduğunu ölçen bir özellik önem derecesi sağlar. Örneğin, karar ağaçları ve rastgele orman gibi algoritmalar, her bir özelliğin ağaçlar üzerindeki etkisini ölçer ve bunu önem sırasına göre sıralar.

Gömülü yöntemlerin avantajları şunlardır:

- Model eğitimi sırasında doğrudan kullanıldıkları için, ayrı bir özellik seçimi adımı gerekmez.
- Modelin kullanacağı özelliklerle birlikte, belirli bir algoritmanın performansını artırmak için en uygun özellikleri belirlerler.
- Özelliklerin etkilerini doğrudan modelin içsel yapısından öğrenirler.

Ancak, gömülü yöntemlerin dezavantajı, bazı algoritmaların özelliklerin önemini belirlemek için daha az duyarlı olabileceği ve belirli türlerde özelliklerin gömülü yöntemlerle doğru bir şekilde öğrenilemeyebileceğidir. Bu nedenle, farklı problem türleri için en uygun özellik seçimi stratejisini belirlemek için dikkatli bir şekilde değerlendirilmelidir.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel


breast_cancer = load_breast_cancer()
x = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
y = breast_cancer.target

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)

scaler = StandardScaler()
scaler.fit(X_train)

selector = SelectFromModel(
    LogisticRegression(C=1000, penalty='l2', max_iter=300, random_state=10))

selector.fit(scaler.transform(X_train), y_train)

X_train_log=selector.transform(X_train)
X_test_log=selector.transform(X_test)
```

```python
breast_cancer = load_breast_cancer()
x = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
y = breast_cancer.target

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)

scaler = StandardScaler()
scaler.fit(X_train)

coefs_df = []

penalties = np.linspace(0.00005,0.5,10)

for c in penalties:   
    
    logit = LogisticRegression(C=c, penalty='l2', random_state=10, max_iter=300)
    logit.fit(scaler.transform(X_train), y_train)
    
    coefs_df.append(pd.Series(logit.coef_.ravel()))
```

```python
coefs = pd.concat(coefs_df, axis=1)
coefs.columns = penalties
coefs.index = X_train.columns
coefs.head()
```

```python
# plot the change in coefficients with the penalty
coefs.plot(figsize=(12,5), legend=True)
plt.xlabel('Penalty value')
plt.ylabel('Coefficient')
plt.title('Coefficient value vs penalty. Each line corresponds to one variable');
```

Katsayılar arasındaki ilişkinin karmaşık olduğunu görebiliriz. Bazı özellikler, cezalandirilmamis haliyle diğerlerinden daha büyük katsayılar gösterir; ancak ceza uygulandığında, bu ilişki değişir. Diğer bazı özellikler, cezalandirilmamis haliyle çok önemlidir (büyük katsayılar gösterir); ancak düzenlenme sonrası önemleri diğer özelliklerinkine benzerdir. Bu durum, özellik seçimi için çok önemli olmasa da, aslında bir değişkenin hedefle olan gerçek ilişkisini anlamak için önemlidir. Başka bir deyişle, Doğrusal Regresyonun sonucunu yorumlamamız ve tam olarak anlamamız gerektiğinde bu detaylar önemlidir.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

breast_cancer = load_breast_cancer()
x = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
y = breast_cancer.target

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)

selector = SelectFromModel(
    RandomForestClassifier(n_estimators=100,random_state=10))

selector.fit(X_train, y_train)

X_train_rf=selector.transform(X_train)
X_test_rf=selector.transform(X_test)
```

#### Yinelemeli Özellik Eleme (Recursive Feature Selection-RFE)

Doğrusal modeller veya karar ağaçlarından elde edilen özelliklerin önemi, veri kümesindeki diğer özelliklerden etkilenir. Dolayısıyla, bazı özelliklerin kaldırılması durumunda, kalan değişkenlerin önemi değişebilir. Bu özellikle, alt kümedeki kalan özelliklerden biriyle korelasyonlu olan özelliklerin kaldırılması durumunda geçerlidir. Kalan özelliğin önemi önemli ölçüde artabilir.
Özellik alt kümelerinin değiştirilmesiyle tetiklenen özellik önemindeki değişikliklere uyum sağlamak için, özellikleri tekrar tekrar kaldırabiliriz. Önem veya katsayısı belirli bir eşiğin altında olan tüm özellikleri kaldırmak yerine, sadece en küçük öneme sahip olanı kaldırırız.
Bu tekrarlı özellik elenmesi uygulamasında, algoritma aşağıdaki gibi işler:

- Bir makine öğrenimi modeli eğitilir.
- Özellik önemini çıkar ve özellikleri sırala.
- En düşük öneme sahip olan özelliği kaldırın.
- Kriter karşılanana kadar 1 ila 3'ü tekrarlayın.

RFE'nin bir dezavantajı, her bir özellik kaldırıldığında ve model yeniden eğitildiğinde hesaplama maliyetinin artmasıdır, özellikle büyük veri setlerinde veya karmaşık modellerde. Bununla birlikte, RFE'nin avantajları, modelin yüksek boyutlu veri setlerinde etkili bir şekilde çalışmasını sağlaması ve daha iyi genelleme yeteneği sağlamasıdır.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC 
from sklearn.feature_selection import RFE

breast_cancer = load_breast_cancer()
x = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
y = breast_cancer.target
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)

selector = RFE(
    RandomForestClassifier(random_state=10))

selector.fit(X_train, y_train)

X_train_rfe=selector.transform(X_train)
X_test_rfe=selector.transform(X_test)
```

```python
selected_feat = X_train.columns[(selector.get_support())]
len(selected_feat)
selector.get_feature_names_out()
selector.estimator_.feature_importances_
selector.feature_names_in_
selector.ranking_
```

```python
results = pd.DataFrame(data= selector.estimator_.feature_importances_, index=selector.get_feature_names_out(), columns=['scores'])
f, ax = plt.subplots(figsize=(12,2), nrows=1, ncols=1)
results.sort_values(by='scores', ascending=False).plot.bar(figsize=(12,2), ax=ax)
ax.set_xlabel('degiskenler')
ax.set_ylabel('RF scores')
ax.set_title('The bigger the better');
```

```python
def run_model(X_train, X_test, y_train, y_test):
    model = SVC(random_state=0)
    model.fit(X_train, y_train)
    
    pred = model.score(X_train,y_train)
    print('Train set: {}'.format(pred))
    
    pred = model.score(X_test,y_test)
    print('Test set: {}'.format(pred))
```

```python
run_model(X_train_log,X_test_log,y_train,y_test)
run_model(X_train_rf,X_test_rf,y_train,y_test)
run_model(X_train_rfe,X_test_rfe,y_train,y_test)
```
