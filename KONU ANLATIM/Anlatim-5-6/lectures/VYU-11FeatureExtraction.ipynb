#### Boyut Azaltma

The state-of-art olarak boyut azaltma teknilerin karsilastirildigi [Understanding How Dimension Reduction Tools Work: An Empirical Approach to Deciphering t-SNE, UMAP, TriMAP, and PaCMAP for Data Visualization](https://arxiv.org/pdf/2012.04456.pdf) calismasini oncelikle incelemenizi oneririz.

- **PCA (Principal Component Analysis - Temel Bileşen Analizi)**

Veri setindeki değişkenliği azaltmak için kullanılan bir boyut azaltma tekniğidir. Veri setindeki değişkenler arasındaki ilişkileri özetlemek ve daha az sayıda değişkenle orijinal veriyi temsil etmek için kullanılır. Bu konu aslında matematikte çok kullanılan "özdeğer", "özvektör", "kovaryans matrisi" gibi kavramları arka planda kullanır. Temel amacı yüksek boyutlu verilerde en yüksek varyans ile veri setini tutmak ancak bunu yaparken boyut indirgemeyi sağlamak olan bir tekniktir. Fazla boyutlu verilerdeki genel özellikleri bularak boyut sayısının azaltılmasını, verinin sıkıştırılmasını sağlar. Boyut azalmasıyla bazı özelliklerin kaybedileceği kesindir; fakat amaçlanan, bu kaybolan özelliklerin popülasyon hakkında çok az bilgi içeriyor olmasıdır. Bu yöntem, yüksek korelasyonlu değişkenleri bir araya getirerek, verilerdeki en çok varyasyonu oluşturan “temel bileşenler” olarak adlandırılan daha az sayıda yapay değişken kümesi oluşturur. 

PCA kullanilirken degiskenler arasinda lineer iliski olmamasi PCA sonucunu olumsuz etkiler. Ayni zamanda verider outlier degerler var ise PCA yaniltici sonuclar verebilir. 

Kac bilesen kullanilmasi gerektigini secilen bilesen sayisina gore verinin varyansinda aciklanabilirlige bakilir. 

Ayrica, veriler icersiinde non-liner bir iliski oldugunuda [Kernel PCA](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.KernelPCA.html) metodu etkilir. 

- **LDA (Linear Discriminant Analysis - Doğrusal Ayırıcı Analiz)**:

   Sınıflar arasındaki farklılıkları en iyi şekilde gösteren özelliklerin bulunması için kullanılan bir boyut azaltma ve özellik seçme tekniğidir. Her bir sınıf içindeki varyansı en aza indirirken sınıflar arasındaki ayrımı en üst düzeye çıkarmayı amaçlar


- **UMAP (Uniform Manifold Approximation and Projection - Düzgün Manifold Yaklaşımı ve Projeksiyonu)**:

   Yüksek boyutlu verileri düşük boyutlu bir uzayda temsil etmek için kullanılan bir boyut azaltma ve görselleştirme tekniğidir. Komşuluk yapısını koruyarak veriyi görselleştirmeyi sağlar. LDA ve PCA algoritmarin ozelliklerini tasimasi ile daha kullanislidir.
   
   [Tutorial](https://pair-code.github.io/understanding-umap/)

- **GaussianRandomProjection**:
  
Secilen n_components (n) parametresi ile $N(0,1/n)$ dagilimndan veriler cekilerek olusturulan bir matrisin araciligi ile yuksek boyutlu bir veri daha dusuk boyutlu hale getirilir. GaussianRandomProjection, boyut azaltma sürecinde rasgele olarak seçilen bir dizi Gauss dağılımından gelen değerlerle oluşturulmuş bir dönüşüm matrisi kullanır. Bu dönüşüm matrisi, yüksek boyutlu verileri daha düşük boyutlu bir altuzayda temsil ederken, veri setindeki varyansı mümkün olduğunca iyi korumaya çalışır. Boyut azaltma işlemi sırasında, veri setindeki özelliklerin sayısı azalırken, olabildiğince az bilgi kaybı yaşanması amaçlanır. Ama her zaman verityi temsil edecek iyi bir rastgele matris olusmayabileceginden bilgi kaybi fazla olabilmektedir.

```python
from sklearn.datasets import load_digits

digits = load_digits()
x=digits.data
y=digits.target
print(x.shape)

pca = PCA().fit(x)
print(np.cumsum(pca.explained_variance_ratio_))

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
plt.show()
print(np.cumsum(pca.explained_variance_ratio_)[20])

pca = PCA(n_components=2).fit(x)
x_pca = pca.transform(x)

sc=plt.scatter(x=x_pca.iloc[:,0], y=x_pca.iloc[:,1], c=y)
plt.legend(handles=sc.legend_elements()[0], labels=list(range(10)))
```

```python
lda=LinearDiscriminantAnalysis(n_components=2)
x_lda=lda.fit_transform(x,y)

sc=plt.scatter(x=x_lda.iloc[:,0], y=x_lda.iloc[:,1], c=y)
plt.legend(handles=sc.legend_elements()[0], labels=list(range(10)))
```
