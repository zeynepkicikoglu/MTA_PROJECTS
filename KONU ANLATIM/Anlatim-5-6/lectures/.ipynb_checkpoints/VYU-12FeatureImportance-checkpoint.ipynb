### Feature Selection

**Note** Ozellik secme kisminda agirlikli olarak Scikit-learn, feature-engine ve mlxtend kutuphaneleri kullanilacak.

Özellik seçimi, makine öğrenme algoritmalarını eğitmek için bir veri setindeki toplam değişkenlerin bir alt kümesini seçme sürecidir. Özellik seçimi teknikleri, değişkenlerin orijinal temsillerini değiştirmez, ancak sadece bunlardan bir alt küme seçerler.
Değişken seçiminin amaçları, makine öğrenme modellerinin performansını iyileştirmek, daha hızlı ve daha maliyet-etkin algoritmalar oluşturmak ve modeller tarafından döndürülen tahminlerin anlaşılmasını kolaylaştırmaktır.
Kisaca, ozellik secme islemleri bizlere asagidakileri saglar:
* Model anlasilmasi kolaylastirir.
* ALgoritmalar daha hizli sonuclanir.
* Bilgisayar bellek problemi cozer.

Ozellik secme algoritmalari yarismalarda nasil kullanildigini gormek icin [Winning the KDD Cup Orange Challenge with Ensemble
Selection](https://proceedings.mlr.press/v7/niculescu09/niculescu09.pdf) calismasini inceleyebiliriz.
Pek cok metod mevcuttur. Bu metotlari
- **Filter Methods**
  Filtre yöntemleri, makine öğrenimi modeliyle etkileşimlerini göz ardı ederek verilerin içsel özelliklerine göre özellikleri seçer. Modelden bağımsız oldukları söylenir. Dolayisiyla, bir kere ozellik secme islemi tamamlandiktan sonra farkli makine ogrenmesi modelleri bu secilmis ozellikler uzerinde performans karsilastirilmasi yapilabilir. Bu surecte, verilerin ozelliklerini belirli algoritmalar ile skor degeleri olusturup sonrasinda belirlenen bir esik degeri ile en yuksek olanlar secilir. Her bir degisken icin egitilmis bir modelin performansina dayanarak degiskenler skorlanabilirken, degiskenin kendi ic ozelliklerine gore de bir skorlama yapilabilir. Cift gozlemler ya da korelasyonu yuksek degiskenler bu filtreleme metotlari icerisinde yer alirlar.
    Diger yontemler ile karsilastirildiginda modelin performansi hususunda pek iyi sayilmasa da hiz acisindan buyuk verilerde avantaj saglamaktadir.
  - Ranking features: ANOVA, chi2

- **Wrapper Methods**
  Sarmalayıcı yöntemler, aramayı tahmine dayalı bir modelin etrafına saran algoritmalardır. Birden fazla özellik alt kümesi oluştururlar ve ardından performanslarını sınıflandırma veya regresyon modeline göre değerlendirirler. Her secilen ozellik alt kumesinin model ile hesaplanmasi epeyce zaman alabilmektedir. MEsela, n tane degiskeni olan bir veride muhtemele butun ozellik altkumeleri dusunuldugunde $2^n$ tane model egitmek gerekeceginden zaman konusunda dezavantaja saglamaktadir. Bu sureci hizlandirmak icin yaygin kullnailan ileri ve geri yonlu arama algoritmalari ile yapilabilir. İleri seçimde, algoritma tüm bireysel özellikleri değerlendirmeye ve en iyisini seçmeye başlar. Daha sonra kalan tüm özellikleri sırayla birer birer ekleyerek inceler. Geriye doğru aramada, algoritma tüm özellik alt kume uzayını inceleyerek başlar ve ardından özellikleri sırayla birer birer kaldırır.

  Bu yontem ile secilen degiskenler, model bazli oldugundan farkli modellerde farkli degiskenler onemli olabilmektedir. Model ile iliskili olmasi avantaj olmasina ragmen hesaplama zamani konusunda dezavantaj saglamaktadir.
  
- **Embedded Methods**
  Gomulu yontemler, modelin egitim asamasinda yani optimizasyon surecinde kullanilir. Wrapper metotlar gibi model ile iliskli bir metottur ancak daha hizli sonuclanmaktadir. Doğrusal regresyon ve karar ağacı tabanlı modeller gibi bazı makine öğrenimi modelleri, optimizasyon prosedürlerinin bir parçası olarak özelliğin önemini belirleyebilir. Lasso regresyon yonteminde onemsiz olan degiskenlerin katsayilari sifirlanarak yapildigindan kolayca aciklanabilirken karar agaclari modelinde degiskenin onemini aciklamak o kadar kolay olmayabilir. Bu metotun dezavanti, her makine ogrenmesi modeli icin uygun olmamasidir.

- **Digerleri**
  Filtre, sarmalayıcı veya gömülü kategorilerinden herhangi birine sınıflandırılması biraz daha zor olan ek özellik seçim prosedürleri vardır. Bu yöntemlerden bazıları, özellik uzayının araştırılmasına rehberlik etmek için gömülü yöntemler gibi bir makine öğrenimi modelinden özellik önemini türetir. Ancak gömülü yöntemlerden farklı olarak, özellik performansını bir makine öğrenimi modeli kullanarak değerlendirirler. Böylece, seçim sürecinde birden fazla modeli eğitirler. Sarmalayıcı yöntemlerden farklı olarak, bu özellik seçim prosedürleri daha az kapsamlı bir özellik alt küme uzayını keşfedecektir. Bu nedenle, çok daha hızlı olma eğilimindedirler. Bu prosedürlere örnek olarak özyinelemeli özellik eleme(Recursive Feature Elimination) veya özyinelemeli özellik ekleme(Recursive Feature Addition) verilebilir.

### Degiskenler arasi Korelasyon
Korelasyon, iki rastgele değişken arasındaki ilişkinin derecesinin istatistiksel bir ölçüsüdür. Korelasyon katsayısı -1 ile 1 arasında değişir. 1 değeri mükemmel bir pozitif korelasyonu, -1 değeri ise mükemmel bir negatif korelasyonu gösterir.

İki değişken birbiriyle ilişkili olduğunda, değişkenlerden birinin değerlerini bilerek ikinci değişkenin değerlerini çıkarabiliriz. Bu değişkenler makine öğrenimi modellerinde tahmin edici olduğunda, değişkenlerden birinin hedefi tahmin etmek için tüm tahmin edici bilgileri zaten sağladığını söyleyebiliriz. Dolayısıyla ikinci değişken gereksizdir.

Korelasyonu tahmin etmek veya bir çift değişken için korelasyon katsayısını hesaplamak için çeşitli yöntemler vardır. Bunlardan en yaygın olanı, iki değişken arasındaki doğrusal ilişkinin derecesini ölçen Pearson korelasyon katsayısıdır.
Spearman'ın sıra korelasyon katsayısı veya Kendall'ın Tau korelasyon katsayısı gibi diğer korelasyon katsayıları değişkenler arasındaki doğrusal olmayan yani monotonic ilişkileri ölçer.

Pearson's korelasyon katsayisi: $\mathrm{X}$ ve $\mathrm{Y}$ rasgele degiskenlerinin lineer iliskileri icin $\bar{X}$ ve $\bar{Y}$ ortalama ve $\sigma_x$ and $\sigma_y$ standart sapmasi olmak uzere 
$$
R=\sum \frac{(X-\bar{X})(Y-\bar{Y})}{\sigma_x \sigma_y}
$$
ile hesaplanir. Spearman korelasyon katsayisi ile Pearson korelasyon formulunu kullanir. Ancak Pearson da degiskenlere ait gozlemlerin degerleri kullanilirken Spearman'da degiskenlerin sirasi kullanilir. Bu siralama isinde, degiskenin sahip oldugu en yuksek degere 1 skoru ve ardindan bir sonraki degere 2 verilerek en kucuk degere kadar devam eder. 

Kendall's tau katsayisi: Degiskenlerin sira degerleri ile hesaplanarak degiskenler arasi monotonic yani dogrusal olmayan iliskleri yakalamak icin asagidaki formulu kullanir. Bu formulde, her iki degiskende de ayni siralamayi paylasan gozlem sayisin  $\mathrm{C}$ ile farkli siralamaya sahip gozlem sayisi da $\mathrm{D}$ ile gosterilir. 

$$
R=\frac{(C-D)}{(C+D)}
$$

Egitim icin linear bir model kullanilacaksa Pearson korelasyon katsayisi kullanilmasi, diger durumda, ornegin karar agaclari gibi bir model kullanilacaksa, Spearman ya da Kendall's Tau kullanilmasi onerilir.

## Filter Metotu

Bu metotlar makine ogrenmesi modeli icermeden degiskenlerin kendi icindeki ozelliklere istatistiksel testler ile skor atamasi yapildiktan sonra siralama yapilir. Bu siralamada en iyi k tanesi ile secim yapilir. Basit ve hizli bir yontem oldugundan tercih edilir.  Istatistiksel testlerden bazilari:
+ $\chi^2$ **Kategorik degiskenler & Kategorik hedef**
    Pearson'ın ki-kare testi kategorik değişkenler arasındaki ilişkiyi belirlemek için kullanılır. Dolayısıyla, bu testi kategorik veriler ile ikili veya çok sınıflı bir hedef arasındaki ilişkiyi değerlendirmek için kullanabiliriz.
    * $H_0$: Iki degisken bagimsiz.
    * $H_1$: Iki degisken bagimsiz degil.
    * $\chi^2$ istatistik formulu: $\chi^2=\sum\frac{(Observed-Expected)^2}{Expected}$
    * Her kategorik degisken icin $\chi^2$ testi yapilip $p-$value degeri bulunduktan sonra her degiskenin ozellik skor degeri olarak $p-$value atanir. Dusuk olan $p-$value degeri, degsiken ile hedef degisken arasinda kuvvettli bir iliskiinin oldugu anlamina gelir. 
+ ANOVA (Analysis of Variance) **Surekli degisken & Kategorik Hedef**
  Hedef değişken ile tahmin edici değişkenler arasında ilişkileri incelemek için kullanılır. Farklı hedef değişkenler, mesela ikili sınıf olduğu durumda, şu soruya cevap verebilmektedir: İki sınıfa ait değişkenlerin ortalaması arasından bir fark var mı? Hipotez testi şu şekilde:
    * $H_0$: Farklı grupların ortalaması eşittir.
    * $H_1$: Farklı grupların ortalaması eşit değildir.
  ANOVA, gruplar ici, varyanslarin esitligini ve gruplarin kendi iclerinde normal daigilima sahip ve gozlemlerin bagimsiz oldugu varsayimina sahiptir. Ayrica, gruplarin kardinalitesi ne kadar birbirine yakin ise ANOVA test sonucu o kadar guvenilirdir.

[Kaynak 2](http://www.stat.columbia.edu/~fwood/Teaching/w4315/Fall2009/lecture_6.pdf)
+ Correlation **Surekli degisken & Surekli hedef**

#### pandas ile chi2 

````python
data = pd.read_csv('../data/feature-engine/titanic.csv')
data.dropna(axis=1, how='any', inplace=True)
data=data[['survived','pclass','sex','sibsp','parch']]
data.head()
````

````python
c=pd.crosstab(index=data['survived'], columns=data['sex'])
c
````

````python
from scipy.stats import chi2_contingency
chi2_contingency(c)
````

````python
def chi2test(x,y):
    c=pd.crosstab(index=x, columns=y)
    _, p_value, _ , _= chi2_contingency(c)
    return p_value
p_values = [chi2test(data['survived'],data[column]) for column in ['pclass', 'sex', 'sibsp', 'parch']]
p_values
````

````python
pd.Series(p_values, index=['pclass', 'sex', 'sibsp', 'parch']).sort_values(ascending=True).plot.bar(rot=45)
````

Buradan yorumlamak biraz ters olabilmektedir. Dusuk p-value degerleri daha kiymetlidir. Yani hedef degisken uzerinde tahminleyici ozellikligi daha fazladir. Bu adimlari asagida okumasi rahat bir sekile donusturebiliriz.

````python
scores = pd.Series(p_values, index=['pclass', 'sex', 'sibsp', 'parch']).sort_values(ascending=True)
scores = -np.log(scores)
scores /= scores.max()
f, ax = plt.subplots()
ax=scores.plot.bar(rot=0)
ax.set_xlabel('degiskenler')
ax.set_ylabel('$-\log(pvalues)$')
plt.show()
````

````python
results = pd.DataFrame(data= p_values, index=['pclass', 'sex', 'sibsp', 'parch'], columns=['pvalues'])
results[['pvalues_log']]=-np.log(results[['pvalues']])
results.replace(to_replace=np.inf, value=2 * np.max(results['pvalues_log'][~np.isinf(results['pvalues_log'])]), inplace=True)
results[['pvalues_log']]/= results[['pvalues_log']].max()
f, ax = plt.subplots(figsize=(12,2), nrows=1, ncols=1)
results[['pvalues_log']].sort_values(by='pvalues_log',ascending=True).plot.bar(rot=40, ax=ax)
ax.set_xlabel('degiskenler')
ax.set_ylabel('$-\log(pvalues)$')
ax.set_title('The bigger the better')
plt.show();
````

## Otomatik Secim
Simdi yukarida yapilan adimlari otomatiklestirerek yapmak istersek, verinin degiskenleri(ozellikleri) arasindan bir puanlama stratejisi(yukaridaki ornekte:chi2) belirlendikten sonra secme islemini en iyi k tane secme islemi yapilabilir. Bazen en iyi k tanesi belirlemek zor olabiliyor bu durumda siralanmis puan listesinde bir yuzdelik dilim icerisine dahil olanlar secilebilir. Istatistiksel testler yapildigi icin secilen $\alpha$ degerine gore de bi esik degeri belirlenebilir. Bu durumlarda  Sklearn kutuphanesinin sagladigi fonksiyonlar kullanilabilir. 

- SelectKBest
- SelectPercentile
- SelectFpr

Daha fazlasi icin [feature_selection](https://scikit-learn.org/stable/modules/feature_selection.html#feature-selection) kismi incelenebilir.

````python
from feature_engine.encoding import OneHotEncoder
data_enc=OneHotEncoder(drop_last_binary=True).fit_transform(data)
data_enc.head()
x=data_enc.drop(['survived'], axis=1)
y=data_enc[['survived']]
````

````python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)

from sklearn.feature_selection import SelectKBest, SelectPercentile, chi2

selector=SelectKBest(score_func=chi2, k=2)
selector.fit(X_train,y_train)
selector.transform(X_train)
````

````python
def plot_scaled_pvalues(selector):
    results = pd.DataFrame(data= np.c_[selector.pvalues_, selector.scores_], index=selector.feature_names_in_, columns=['pvalues', 'scores'])
    results[['pvalues_log']]=-np.log(results[['pvalues']])
    results.replace(to_replace=np.inf, value=10 * np.max(results['pvalues_log'][~np.isinf(results['pvalues_log'])]), inplace=True)
    results[['pvalues_log']]/= results[['pvalues_log']].max()
    f, ax = plt.subplots(figsize=(12,2), nrows=1, ncols=1)
    results[['pvalues_log']].sort_values(by='pvalues_log',ascending=True).plot.bar(rot=40, ax=ax)
    ax.set_xlabel('degiskenler')
    ax.set_ylabel('$-\log(pvalues)$')
    ax.set_title('The bigger the better')
    plt.show();
````

## Mutual Information
**Bilgi Teorisi (Information Theory)**
Istatistik ve makine ogreniminde siklikla kullanilan bilgi teoresi kavramlarida ozellik secmede puanlama kisminda kullanilabilir. 

**Entropi**, bir rastgele değişkenin belirsizliğini ölçen bir kavramdır. Bir rastgele değişkenin entropisi, o değişkenin olası değerlerinin dağılımına dayanır. Daha yüksek entropi, daha fazla belirsizlik anlamına gelir. Bir rastgele değişkende beklenen "bilgi miktarını" nicelendirir. Entropi, bilgi teorisinde önemli bir rol oynar ve rastgele değişkenlerin tahmin edilmesi, sıkıştırılması ve kodlanması gibi birçok uygulamada kullanılır. Örneğin, bir veri kümesinde iki sınıf varsa ve her iki sınıf da eşit sayıda örneğe sahipse, entropi en yüksektir çünkü belirsizlik en fazladır. bir rasgele degsikenin entropi deger asagidaki formul ile hesaplanir:

   $$H(X) = -\sum_{x \in \mathcal{X}} p(x) \log_2(p(x))$$

Burada, $X$ bir rastgele değişken ve $p(x_i)$, $X$'in $i$'inci değerinin olasılığıdır.

**Gini katsayısı**, bir olasılık dağılımının eşitsizlik ölçüsünü ifade eder. Özellikle sınıflandırma problemlerinde kullanılır. Gini katsayısı, bir rastgele değişkenin değerlerinin dağılımının ne kadar eşit veya eşitsiz olduğunu ölçer. Değerleri 0 ile 1 arasında değişir ve 0, tamamen eşit dağılımı, 1 ise tamamen eşitsiz dağılımı temsil eder. Sınıflandırma ağaçları gibi algoritmalar, Gini katsayısını kullanarak en iyi bölünmeyi seçerler. Gini katsayısı, aşağıdaki formülle hesaplanır:

$$Gini(X) = 1 - \sum_{x \in \mathcal{X}} p(x)^2$$

Burada, $X$ bir rastgele değişken ve $p(x_i)$, $X$'in $i$'inci değerinin olasılığıdır.

**Kullback–Leibler Divergence (KL Divergence)**: KL divergence, iki olasılık dağılımı arasındaki farkı ölçer. Bir veri kümesinin gerçek dağılımı $P$ ile bir referans dağılımı $Q$ arasındaki farkı nicelendirir. KL divergansı aşağıdaki formülle hesaplanır:

   $$D_{KL}(P \| Q) = \sum_{x \in \mathcal{X}} p(x) \log\left(\frac{p(x)}{q(x)}\right)$$

   Burada, $p(x_i)$ ve $q(x_i)$ sırasıyla $P$ ve $Q$'nun $i$'inci değerlerinin olasılıklarıdır.

- KL ayrışması her zaman 0 veya daha büyüktür.
- KL ayrışması 0 ise, P ve Q dağılımları aynıdır.
- KL ayrışması ne kadar büyükse, P ve Q dağılımları arasındaki fark o kadar büyüktür.


[Kullback–Leibler divergence - Wikipedia](https://en.wikipedia.org/wiki/Kullback%E2%80%93Leibler_divergence.)\
[Kullback-Leibler Divergence - University of Illinois Urbana-Champaign.](http://hanj.cs.illinois.edu/cs412/bk3/KL-divergence.pdf.)

[**Mutual information (ortak(karsilikli) bilgi)**]((https://en.wikipedia.org/wiki/Mutual_information)), iki rastgele değişken arasındaki bağımlılık miktarını ölçen bir bilgi teorisidir. Basitçe, bir değişkenin değerini bilmenin diğer değişkenin belirsizliği üzerinde ne kadar etkiye sahip olduğunu gösterir.

Daha spesifik olarak, bir rastgele değişkeni gözlemleyerek diğer rastgele değişken hakkında ne kadar "bilgi" elde edildiğini (shannonlar (bit), natlar veya hartleyler gibi birimlerde) nicelendirir. Mutual information kavramı, rastgele bir değişkenin entropisi ile sıkı bir şekilde bağlantılıdır. MI, gerçek değerli rastgele değişkenlerle ve doğrusal bağımlılıkla sınırlı değildir (örneğin korelasyon katsayısı gibi). MI daha geneldir ve çiftin ortak dağılımının ve marginals dağılımlarının çarpımından ne kadar farklı olduğunu belirler. 

İki rasgele değişken $X$ ve $Y$ arasındaki karşılıklı bilgi, ortak dağılımları ile marjinallerinin çarpımı arasındaki KL farklılığı olarak da tanımlanabilir.

MI Hesaplama:

1) Iki tane kesiklik rasgele degisken  $X$ ve $Y$ arasinda mutual information asagidaki formul ile hesaplanir:

$$\mathrm{I}(X, Y)=-\sum_{x \in \mathcal{X}} \sum_{x \in \mathcal{Y}} p(x, y) \cdot \log \frac{p(x, y)}{p(x) \cdot p(y)}$$

+ $p(x)$ ve $p(y)$: olasılık kütle fonksiyonları (the probability mass functions of $X$ and $Y$)
+ $p(x, y)$: bilesik olasik kutle fonksiyonu (the joint probability mass function of $X$ and $Y$).

2) Iki tane surekli rasgele degisken  $X$ ve $Y$ arasinda mutual information asagidaki formul ile hesaplanir: 
$$\mathrm{I}(X, Y)=-\int_{\mathcal{X}} \int_{\mathcal{Y}} p(x, y) \cdot \log \frac{p(x, y)}{p(x) \cdot p(y)} \mathrm{d} y \mathrm{~d} x$$

+ $p(x)$ ve $p(y)$: olasılık yogunluk fonksiyonları (he probability density functionsof $X$ and $Y$)
+ $p(x, y)$: bilesik olasik yogunluk fonksiyonu (the joint probability density function of $X$ and $Y$).


MI Özellikleri:

- MI, 0 ile $\infty$ arasında bir değer alır.
- MI değeri 0 ise, X ve Y arasında bağımlılık yoktur.
- MI değeri $\infty$ 'e yaklaşıyorsa, X ve Y arasında tam bağımlılık vardır.


#### Feature Selection'da Mutual Information

Feature selection (özellik seçimi), bir makine öğrenimi modelinin performansını ve sadeliğini optimize etmek için en önemli özellikleri seçme işlemidir. MI, feature selection'da aşağıdaki gibi kullanılabilir:

1. Feature Ranking:

Her bir özelliğin hedef değişkenle olan MI'si hesaplanarak, en yüksek MI'ye sahip özellikler seçilebilir.

2. Feature Subset Selection:

Farklı özellik alt kümelerinin MI'leri hesaplanarak, en yüksek MI'ye sahip ve aynı zamanda birbirleriyle düşük MI'ye sahip alt küme seçilebilir.

- Mutual Information'ın Feature Selection'da Kullanımının Avantajları:

+ Hesaplaması kolaydır.
+ Herhangi bir veri dağılımı varsayımı yapmaz.
+ Hem sayısal hem de kategorik özelliklerle kullanılabilir.

- Mutual Information'ın Feature Selection'da Kullanımının Dezavantajları:
  
+ Özellikler arasındaki doğrusal olmayan ilişkileri yakalayamaz.
+ Gürültüye karşı hassastır.

[Kaynak](http://www.stat.columbia.edu/~liam/teaching/neurostat-fall17/papers/EM/Cover&Thomas-Ch2.pdf)
[Kaynak](https://machinelearningmastery.com/information-gain-and-mutual-information/)
[Kaynak](https://hanj.cs.illinois.edu/cs412/bk3/KL-divergence.pdf)

````python
from sklearn.feature_selection import mutual_info_classif, SelectKBest
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
x = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
y = breast_cancer.target

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)

selector=SelectKBest(score_func=mutual_info_classif, k=5)
selector.fit(X_train,y_train)
selector.transform(X_train)
````

````python
results = pd.DataFrame(data= selector.scores_, index=selector.feature_names_in_, columns=['scores'])
f, ax = plt.subplots(figsize=(12,2), nrows=1, ncols=1)
results.sort_values(by='scores', ascending=False).plot.bar(figsize=(12,2), ax=ax)
ax.set_xlabel('degiskenler')
ax.set_ylabel('MI degerleri')
ax.set_title('The bigger the better');
````
