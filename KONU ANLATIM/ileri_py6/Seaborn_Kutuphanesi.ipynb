# Seaborn Kütüphanesi 

Seaborn Kütüphanesi: Matplotlib temel alınarak oluşturulmuş veri görselleştirme kütüphanesidir.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Histogram Grafiği

```python
data = np.random.normal(0, 1, 100)
sns.histplot(data)
plt.show()
```

### Saçılım Grafiği

```python
data = sns.load_dataset('iris') #İris datasını yükle

sns.scatterplot(x="petal_width", y="petal_length", data=data) #Scatter plot
```

### Box Grafiği

```python
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=tips)
```
