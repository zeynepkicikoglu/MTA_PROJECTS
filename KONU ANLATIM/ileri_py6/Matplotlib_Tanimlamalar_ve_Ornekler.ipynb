### Scatter Plot

```python
## Scatter plot çizdirmek için x ekseni ve y ekseni değerlerinin ikisi de verilmelidir.

# x ekseni verisini verebilmek için views ile aynı uzunlukta bir array oluşturulaım
x_eksen_array = np.arange(0, len(veri_seti["views"]))

plt.figure(figsize = (20,8))
plt.scatter(x_eksen_array, veri_seti["views"],color = "green", label = "Görüntülenme Sayıları")
plt.title("Views")
plt.xlabel("Index Sayısı")
plt.ylabel("Görüntülenme Sayıları")
plt.legend()
plt.grid(True)
```

### Grafiği Kayıt Edelim

```python
# Çizdirdiğimiz grafiği kaydetmek istersek

zaman_arr = np.arange(0.0, 2.0, 0.01) # 0'dan başla 2'ye kadar 0.01'er artarak git
s = 1 + np.sin(2*np.pi*zaman_arr) 
plt.plot(zaman_arr, s)

plt.xlabel('Zaman (s)')
plt.ylabel('Voltaj (mV)')
plt.title('Voltaj Zaman Grafiği')
plt.grid(True)
plt.savefig("Kayıt_Testi.png")
plt.show()
```

### Bar Plot Çizdirelim

```python
# Verisetini Oluşturalım
veri  = {'C':20, 'C++':15, 'Java':30, 'Python':35}

programlama_kurslari = list(veri.keys())
kayitli_kisiler      = list(veri.values())
  
fig = plt.figure(figsize = (10, 5))
 
# Bar plotu oluşturalım
plt.bar(programlama_kurslari, kayitli_kisiler, color ='maroon', width = 0.4)
 
plt.xlabel("Kayit Olunabilecek Kurslar")
plt.ylabel("Kurslara Kayitli Öğrenci Sayisi")
plt.title("Farklı Kurslara Kayit Olmuş Öğrenciler")
plt.show()
```

```python
# Bar Kalınlıklarını Belirleyelim
barKalinlik = 0.25
fig = plt.subplots(figsize =(12, 8)) # Subplots ile birden fazla plot yapabiliriz
 
# Derslerden Geçen Öğrenci Sayıları
Tarih     = [12, 30, 1, 8, 22] 
Matematik = [28, 6, 16, 5, 10] 
Edebiyat  = [29, 3, 24, 25, 17] 
 
# X-ekseninde barların pozisyonlarını ayarlayalım
br1 = np.arange(len(Tarih)) 
br2 = [x + barKalinlik for x in br1] 
br3 = [x + barKalinlik for x in br2] 
 
# Make the plot
plt.bar(br1, Tarih,     color ='r',  width = barKalinlik,   label ='Tarih') 
plt.bar(br2, Matematik, color ='g',  width = barKalinlik,   label ='Matematik') 
plt.bar(br3, Edebiyat,  color ='b',  width = barKalinlik,   label ='Edebiyat') 
 
# Adding Xticks 
plt.xlabel('Ders', fontweight ='bold', fontsize = 15) 
plt.ylabel('Geçen Öğrenci Sayısı', fontweight ='bold', fontsize = 15) 
plt.xticks([r + barKalinlik for r in range(len(Tarih))], ['2015', '2016', '2017', '2018', '2019'])
 
plt.legend()
plt.show() 
```

```python
N = 5
 
ege_uni         = (20, 35, 30, 35, 27)
sakarya_uni     = (25, 32, 34, 20, 25)
indeks          = np.arange(N)   
kalinlik        = 0.35 
 
fig = plt.subplots(figsize =(10, 7))
p1 = plt.bar(indeks, ege_uni, kalinlik)
p2 = plt.bar(indeks, sakarya_uni, kalinlik, bottom = ege_uni)
 
plt.ylabel('Öğrenci Sayıları')
plt.title('Takımlara Göre Sayılar')
plt.xticks(indeks, ('Takım1', 'Takım2', 'Takım3', 'Takım4', 'Takım5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Ege', 'Sakarya'))
 
plt.show()
```

### 3D Grafik Çizdirelim

```python
from mpl_toolkits import mplot3d
 
fig = plt.figure()
 
# 3-D projeksiyon için
ax = plt.axes(projection ='3d')
 
# Kullanacağımız 3 ekseni de tanımlayalım
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
 
ax.plot3D(x, y, z, 'green')
ax.set_title('3D Çizgi Grafiği')
plt.show()
```

```python
fig = plt.figure()
 
# 3-D projeksiyon için
ax = plt.axes(projection ='3d')
 
# Kullanacağımız 3 ekseni de tanımlayalım
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
c = x + y
ax.scatter(x, y, z, c = x+y)
 
ax.set_title('3D Scatter Plot')
plt.show()
```

```python
# Yüzeyleri ve eksenleri tanımlayalım
x = np.outer(np.linspace(-2, 2, 10), np.ones(10))
y = x.copy().T
z = np.cos(x ** 2 + y ** 3)
 
fig = plt.figure()
 
# 3D projeksiyon için
ax = plt.axes(projection='3d')
 
# syntax for plotting
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='green')
ax.set_title('Yüzey Plot')
plt.show()
```
