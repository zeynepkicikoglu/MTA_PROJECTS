**Yapılandırılmış veri örnekleri şunlardır:**

### Sürekli Veri: Aralık ve Oran Verisi (Interval and Ratio Continous Data)

Sürekli veri, genellikle zaman, mesafe ve sıcaklık gibi sürekli bir ölçek üzerine yerleştirilmiş sayısal değerleri içeren bir veri türüdür. Bu değerler genellikle bir aralık veya oran ölçeğinde ölçülür ve belirli bir aralık içinde sonsuz sayıda olası değeri alabilir.

Sürekli veri, aralık ve oran verisi olmak üzere iki türe ayrılabilir. Her iki tür de sıralanmış, eşit ve anlamlı bir şekilde aralıklı değerlerden oluşur. Aralarındaki temel fark, oran verisinin anlamlı bir sıfır noktasına sahip olmasıdır, bu da değerlerin eklenmesine veya çıkarılmasına izin verirken, aralık verisinin anlamlı bir sıfır noktasına sahip olmamasıdır. Oran verisinin örnekleri arasında gelir ve satış miktarı bulunurken, aralık verisinin örnekleri tarihler, pH ölçümleri ve fırın sıcaklıklarıdır.

#### Aralık Verisi (Interval Data)

Aralık verisi, anlamlı ve tutarlı farkları olan sıralanmış değerlerden oluşur; ancak anlamlı bir sıfır noktası yoktur. Başka bir deyişle, aralık verisinin sıfır değeri, özelliğin yokluğunu değil, ölçek üzerinde rastgele bir noktayı gösterir. Aralık verisinin örnekleri şunlardır:

- Celsius veya Fahrenheit cinsinden ölçülen sıcaklık (örneğin, 20°C ile 25°C arasındaki fark, 30°C ile 35°C arasındaki farkla aynıdır, ancak 0°C sıcaklığın yokluğunu göstermez)
- Takvim tarihleri (örneğin, 1 Ocak 2000 ile 1 Ocak 2010 arasındaki fark, 1 Haziran 2015 ile 1 Haziran 2025 arasındaki farkla aynıdır)

Aralık verisi, anlamlı bir sıfır noktasına sahip olmadığı için toplama veya çıkarma gibi işlemler gerçekleştirilebilir, ancak çarpma veya bölme anlamlı değildir ve değerler arasındaki oranlar hesaplanamaz veya yorumlanamaz.Örneğin, 100 derecenin 50 dereceden iki kat daha sıcak olmadığını, aynı şekilde 20 derecenin 40 dereceden iki kat daha soğuk olmadığını biliyoruz. Bu tür şeyleri söylemeyiz çünkü gerçekte bir anlam taşımazlar.

Sayıları ekleyip çıkartma yeteneğine sahip olduğumuzda ve bu eklemelerin veya çıkarmaların tutarlı olduğuna güvenebildiğimizde, aritmetik ortalamalar, medianlar ve standart sapmaları hesaplamaya başlayabiliriz. Bu formüller, değerleri bir araya ekleyebilme yeteneğine ve bu cevapların bir anlam taşımasına dayanır.

**Örnek: Aralık Seviyesinde eksik bir veri ile uğraşma**

Aralık seviyesinde veri ile, eksik verilere sahip olsak bile birçok işlem yapabiliriz. Örneğin, eğer eksik değerlerimiz varsa, eksik verileri doldurmak için aritmetik ortalama veya medyan kullanma gücümüz vardır. İstenirse özellikleri birbirine ekleyip çıkartmaya başlayabiliriz.

Aritmetik ortalama, tüm birimleri (örneğin, ft, ml, USD, vb.) aynı olan verileri birleştirdiğimizde kullanışlıdır. Ancak aritmetik ortalamanın bir dezavantajı, aykırı değerlerden ciddi şekilde etkilenmesidir. Örneğin, 1–100 arasındaki anket puanlarının ortalamasını aldığımızı düşünelim ve puanlarımız şu şekilde olsun: 30, 54, 34, 54, 36, 44, 23, 93, 100, 99. Aritmetik ortalamamız 56.7 olacaktır, ancak medyanımız 49 olacaktır. 90'larda ve 100'lerdeki üç aykırı değerimiz, ortalama değerimizi yapay bir şekilde yukarı çekerken, medyanımız daha çok ortada kalır. Bu durumda veriyi temsilen median değerleri doldurmak makul olacaktır.

Unutmayın ki anket verilerini aralık seviyesinde ele alıyoruz, bu nedenle farklara anlam katmış oluyoruz. Birisinin 95'te olduğunu söyleyerek, 85'te olan birinden yaklaşık 10 birim daha mutlu olduğunu ifade etmekten çekinmiyoruz.

#### Oran Verisi (Ratio Data)

Oran verisi, aralık verisiyle benzer özelliklere sahiptir ve sıralanmış değerler ve aralarındaki tutarlı farklar içerir. Ancak aralık verisinin aksine, oran verisi ölçülen özelliğin yokluğunu gösteren anlamlı bir sıfır noktasına sahiptir. Bu sıfır noktası, çarpma ve bölme de dahil olmak üzere tüm aritmetik işlemleri gerçekleştirmeye olanak tanır. Oran seviyesindeki veri oldukça yaygındır ve örnekleri şunlardır:

- Kelvin cinsinden ölçülen sıcaklık (örneğin, 100 K, 50 K'nın iki katıdır, çünkü 0 K sıcaklığın yokluğunu gösterir)
- Mesafe (örneğin, 10 metre, 5 metrenin iki katıdır)
- Yaş (örneğin, 40 yaşındaki bir birey, 20 yaşındaki birinin iki katıdır)
- Para (Gerçek sıfırı, paranın yokluğunu temsil eden bir durumu tanımlayabiliriz. Eğer 0 dolar veya 0 liramız varsa, hiç para yok demektir.)

**Örnek: Oran Seviyesinde bir veri ile uğraşma**


Aralık seviyesinde veriler ile yapabildiğimizden çok daha fazlası ile oran seviyesinde veriler ile çalışabilmekteyiz. Daha önce, aralık seviyesindeki verilerden bahsederken aritmetik ortalama hakkında konuştuk. Oran seviyesinde, aynı amaçlar için hala aritmetik ortalamayı kullanabiliriz, ancak bazı durumlarda geometrik veya harmonik ortalama kullanmayı tercih etmemiz gerekbilir. Geometrik ortalama, verilerimiz farklı birimlerde (örneğin, Celsius ve Fahrenheit karışımı) veya farklı ölçeklerde ise en kullanışlı olanıdır.

Makine öğreniminde harmonik ortalama kullanımının yaygın bir örneği, F-Score (ayrıca F1-Score veya Fbeta-Score) hesaplanmasıdır; bu, precision ve recall ölçütlerinin harmonik ortalaması olarak hesaplanan bir sınıflandırma model değerlendirme ölçütüdür.

$$\begin{equation}
\text{Aritmetik Ortalama}=\frac{1}{n} \cdot \sum_{i=1}^n a_i \quad \text{Geometrik Ortalama}=\left(\prod_{i=1}^n a_i\right)^{\frac{1}{n}} \quad\text{Harmonik Ortalama}=\left(\frac{1}{n} \cdot \sum_{i=1}^n a_i^{-1}\right)^{-1}
\end{equation}$$

| Tür Ortalaması | Açıklama | Veri Seviyesi | Kullanım Koşulları | Kullanılmaması Gereken Durumlar |
|-----------------|----------|---------------|---------------------|--------------------------------|
| Aritmetik Ortalama | Verilerin tutarlı birimlere sahip olduğu durumlar | Aralık ve Oran | - Veriler tutarlı birimlere sahipse. <br/> - Veriler doğasında toplamaya uygundur. | - Aykırı değerlerin ortalama üzerinde etkilenmemesi gerektiğinde. |
| Geometrik Ortalama | Verilerin farklı ölçeklerde veya birimlerde olduğu durumlar | Oran | - Veriler çarpma işlemine uygunsa. <br/> - Ölçekler ve birimler önemli değilse. | - Ölçekler ve birimler önemliyse; geometrik ortalama bunları gizleyebilir. |
| Harmonik Ortalama | Verilerin oranlara sahip olduğu durumlar | Oran | - Veriler iki değerin bir kesir olarak oranıdır. | - Veriler sıfır veya negatif değerlere sahipse. <br/> - Harmonik ortalama hakkında bilgi sahibi olmayanlar için zor yorumlanabilir. |

#### Sürekli (Aralık ve Oran) Veri İçin Analiz ve Görselleştirmeler

Sürekli veriyi analiz etmek genellikle merkezi eğilim (ortalama, medyan, mod) ve dağılım (aralık (range), çeyrek aralık(interquartile range), varyans, standart sapma) ölçümlerini hesaplamayı içerir. Çeşitli görselleştirme teknikleri, sürekli veriyi temsil etmek için kullanılabilir:

**Histogramlar**: Bir histogram, sürekli verinin frekans dağılımını belirli aralıklara bölen bir bar grafiğidir. Kategorik ve sıralı veriler için kullanılan çubuk grafiklerinden(bar charts) farklı olarak, histogramlarda çubuklar arasında boşluk yoktur ve verinin sürekli doğasını vurgular.

**Çizgi grafikleri(Line Plots)**: Çizgi grafikleri, sürekli verideki eğilimleri veya desenleri tasvir etmek için kullanışlıdır: özellikle zaman serisi verileri bağlamında (örneğin, birkaç gün boyunca sıcaklık trendini gösterme). Veri noktaları, değerlerin sürekli bir ölçekteki ilerlemesini gösteren çizgilerle birbirine bağlanır.

**Kutu grafikleri(Box Plots)**: Kutu grafikleri sürekli verinin merkezi eğilimi, dağılımı ve olası aykırı değerleri görsel olarak temsil eder. Ortanca, alt çeyrek, üst çeyrek ve olası aşırı değerleri (outliers) göstererek veri dağılımının genel bir bakışını sunar.

**Dağılım grafikleri(Scatter Plots)**: Dağılım grafikleri, iki sürekli değişken arasındaki ilişkiyi, veri noktalarını bir koordinat düzlemine yerleştirerek gösterir. Bu tür bir görselleştirme, çizilen değişkenler arasındaki desenleri veya ilişkileri ortaya çıkarmaya yardımcı olur.

#### Sürekli Verinin Özellik Mühendisliğinde İşleme

Sürekli veri genellikle makine öğrenimi modellerine veya istatistiksel analizlere entegre edilmeden önce bazı ön işleme adımlarını gerektirir:

**Ölçekleme(Scaling)**: Ölçekleme, sürekli verinin aralığını ayarlamayı ifade eder ve bu, büyük ölçek, birimler veya aralıklara sahip değişkenlerle başa çıkmaya yardımcı olur. Yaygın ölçekleme teknikleri arasında min-max ölçekleme ve standartlaştırma (z-puanı normalleştirme) bulunur.

**Normalizasyon(Normalization)**: Bu, verinin genellikle 0 ila 1 arasında değişen ortak bir ölçeği takip etmesini sağlamak amacıyla yapılır ve makine öğrenimi algoritmalarının istikrarını ve yakınsamasını artırır.

**Dönüşüm(Transformation)**: Sürekli veriye dönüşümler uygulamak, logaritmik veya üssel dönüşümler gibi, değişkenliği stabilize etmeye, normaliteyi sağlamaya veya bazı modellerin varsayımlarını karşılamak için çarpıcı ilişkileri aditif olanlara dönüştürmeye yardımcı olabilir.

**Özellik seçimi(Feature Selection)**: Yüksek sayıda sürekli özellikle uğraşıyorsanız, özellik seçimi tekniklerini uygulamak uygun olabilir. Bu teknikler arasında korelasyon analizi, temel bileşen analizi (PCA) veya özyinelemeli özellik elemesi(Recursive Feature Elimination) gibi, en önemli tahmincileri belirlemek ve boyutu azaltmak için kullanılabilir.

### Siralı Veriler (Ordinal Data)

Sıralı veriler, değerlerin doğal bir sıra veya derecelendirme sistemine sahip olduğu, ancak ardışık değerler arasındaki farkların anlamlı veya sabit olmadığı bir veri türüdür. Bu veri değerleri, her iki türün bazı özelliklerini paylaştıkları kategorik ve sürekli veriler arasında yer alır. Sıralı veriler, tamsayılar kullanılarak temsil edilebilir, ancak sürekli verilerden farklı olarak toplama veya çıkarma gibi anlamlı aritmetik işlemlere sahip değildir.

Sıralı verilerin ortak örnekleri şunları içerir:

- Likert ölçekleri (örneğin, kuvvetle katılıyorum, katılıyorum, nötr, katılmıyorum, kuvvetle katılmıyorum)
- Eğitim düzeyleri (örneğin, lise, üniversite, yüksek lisans, doktora)
- Sosyoekonomik statü (örneğin, düşük, orta, yüksek)
- Anket yanıtları (örneğin, kötü, orta, iyi, mükemmel)

#### Sıralı Verilerin Analizi ve Görselleştirmeleri

Sıralı verileri analiz etmek genellikle frekans dağılımını, merkezi eğilimi (medyan veya mode gibi) ve yayılım ölçümlerini ( açıklık(range) veya çeyreklik açıklığı(interquartile range) gibi) incelemeyi içerir. Sürekli verilerin aksine, sıralı veriler ortalama veya standart sapma gibi ölçümler kullanılarak analiz edilemez. Sıralı veriler için yaygın olarak kullanılan üç görselleştirme tekniği vardır:

**Sütun Grafikleri(Bar Charts)**: Kategorik veriler gibi, sıralı veriler için de her kategorinin frekansını veya oranını görüntülemek için sütun grafikleri kullanılabilir. Ancak, değerlerin doğal derecelendirmesini temsil etmek için kategori sıralaması x-ekseninde korunmalıdır.

**Çizgi Grafikleri(Line Charts)**: Çizgi grafikleri, sıralı kategoriler arasında eğilimleri veya desenleri görselleştirmek için kullanılabilir ve veri noktalarını çizgi segmentleri ile sıralı bir şekilde bağlayabilir. Kategorilerin sıralamasını koruyarak ve ilerlemeyi göstererek, çizgi grafikleri sütun grafiklerine göre daha fazla bilgi aktarabilir.

**Kutu Grafikleri(Box Plots)**: Bu grafikler, sıralı veriler için merkezi eğilim, yayılım ve olası aykırı değerlerin görsel bir temsilini sunar. Kutu grafikleri, veri dağılımına genel bir bakış sağlamak için median, alt çeyrek, üst çeyrek ve olası ekstrem değerleri (aykırı değerleri) görüntüler.

#### Feature Engineering'de Sıralı Verilerle Çalışma

Makine öğrenimi algoritmaları veya sayısal girdi gerektiren istatistiksel modeller kullanırken, sıralı verileri doğru bir şekilde hazırlamak ve kodlamak önemlidir. Sıralı verileri özellik mühendisliği'nde ele almak için aşağıdaki teknikler kullanılabilir:

**Tamsayı Kodlama(Integer Encoding)**: Bu yöntem, sıralı verileri kodlamak için yaygın olarak kullanilir ve dogal sirayı koruyan bir sekilde kategorilere tamsayı degerleri atamasi ile gerceklestirilir. Örneğin, lise, üniversite, yüksek lisans ve doktora gibi eğitim düzeylerine sırasıyla 1, 2, 3 ve 4 degerleri atanabilir.

**Özel Kodlama(Custom Encoding)**: Bazı durumlarda, alan bilgisine veya problem gereksinimlerine dayalı özel kodlama uygulamak gerekebilir. Örneğin, modele önemleri vurgulamak için sıralı kategorilere ağırlık atayabilirsiniz.

**Binleme veya Ayrıklaştırma(Binning or Discretization)**: Bazen sıralı veriler, kategorilere aralık değerleri atanarak sürekli veriler olarak ele alınabilir. Bu işlem, verilerin eşit aralıklara sahip kutulara veya aralıklara bölünmesini içerir. Bu yaklaşım, bilgi kaybına yol açabileceğinden ve eşit aralıklar varsayımı uygun olmayabileceğinden dikkatle değerlendirilmelidir.

### 3. Kategorik Veri

Kategorik veri, her bir öğenin tek bir kategoriye ait olduğu belirli değerlerden oluşur. Bu tür verilerin doğal bir sıralaması yoktur. Kategorik verilerin etkili bir şekilde görselleştirilebilmesi için, kategorik verilerin benzersiz değerlerin sayısı olan kardinalitesini bilmek önemlidir.

Kategorik veri, aynı zamanda nitel veya nominal veri olarak da bilinir ve sayısal olmayan bir veri türüdür. Bu veri türü, belirli kategorileri veya grupları temsil eden değerlerden oluşur. Her gözlem veya öğe yalnızca bir kategoriye aittir. Sürekli veya sıralı verilerin aksine, kategorik verilerin doğal bir sırası veya sıralaması yoktur. Yaygın örnekler arasında cinsiyet (erkek, kadın), renkler (kırmızı, mavi, yeşil) ve ülkeler (ABD, Fransa, Almanya) bulunur.

#### Temel Kategorik Veri Türleri:

1. **Nominal veri**: Bu tür kategorik veri, doğal bir sıralama veya sıralama sistemine sahip değildir. Örneğin, renkler durumunda, mantıksal olarak kırmızının mavi öncesinde veya sonrasında geldiğini söyleyemezsiniz. Nominal veri, temelde anlamlı bir sayısal temsil olmadan adlandırılmış kategorilerden oluşur.
2. **İkili veri(Binary)**: İkili veri, yalnızca iki farklı kategoriye sahip olan nominal verinin özel bir durumudur. İkili veri örnekleri, kazanç/kayıp sonuçları, bir özelliğin varlığı/yokluğu veya geçme/kalma sonuçlarıdır.

#### Kategorik Veri İçin Analiz ve Görselleştirmeler

Kategorik verilerin analizi genellikle her kategorinin frekansını veya oranını anlamayı içerir. Örneğin, belirli bir ürünü tercih eden bireylerin yüzdesini veya bir nüfus örneğinde dişi ve erkek oranını bilmek isteyebilirsiniz. Kategorik veriler için en yaygın görselleştirme teknikleri arasında çubuk grafikleri, pasta grafikleri ve yığılmış çubuk grafikleri bulunur.

**Çubuk grafikleri(Bar Charts)**: Bu, her kategorinin frekansını veya oranını görselleştirmek için kullanışlıdır. Kategorileri x-ekseninde ve frekansı veya oranı y-ekseninde temsil eder.

**Pasta grafikleri(Pie Charts)**: Her bir kategoriyi bir dairenin dilimi olarak temsil eder, dilimin boyutu o kategorinin frekansına veya oranına dayanır. Pasta grafikleri, farklı kategorilerin oranlarını bir bütün olarak göstermek istediğinizde kullanışlıdır.

**Yığılmış çubuk grafikleri(Stacked Bar Charts)**: Bu, çubuk grafiklerine benzer, ancak birden çok kategoriyi tek bir çubuk içinde segmentler olarak gösterir. Yığılmış çubuk grafikleri, birden çok kategorik değişkeni karşılaştırırken kullanışlı olabilir.

#### Feature Engineering'de kategorik veri ile uğraşma

Birçok makine öğrenimi algoritması ve istatistiksel modelde, kategorik verilerin sayısal bir formata dönüştürülmesi gereklidir. Kategorik verilerle başa çıkma konusunda özellik mühendisliğinde kullanılan çeşitli teknikler bulunmaktadır:

1. **Etiket kodlama (Label encoding)**: Bu yöntem, bir özniteliğin içindeki her kategoriye bir tamsayı değeri atamayı içerir. Örneğin, "kırmızı" için 0, "mavi" için 1 ve "yeşil" için 2 gibi değerler atayabilirsiniz. Ancak, bu yaklaşım, kategoriler arasında düzenli bir ilişki olduğunu varsayar ve bu, nominal veriler için uygun olmayabilir.
   
2. **One-hot encoding**: Bu yaklaşımda, her kategori ayrı bir ikili (0 veya 1) değişken olarak temsil edilir. Örneğin, "kırmızı", "mavi" ve "yeşil" renkleri üç ayrı özellik olarak dönüştürülür, 1 renk varlığını temsil ederken 0 renk yokluğunu temsil eder.

3. **Dummy encoding**: One-hot encoding'e benzer şekilde, dummy encoding kategorileri ayrı ikili değişkenlerle temsil eder. Ancak, çoklu doğrusallık sorunlarını önlemek için bir kategori referans kategori olarak hariç bırakılır. Örneğin, renkler için "kırmızı" ve "mavi" için ikili değişkenler oluşturabilir, her ikisi de yoksa "yeşil" kategorisini temsil eder. Bu teknik genellikle regresyon modellerinde kullanılır.

Doğru kodlama yöntemini seçmek, kategorik değişkenin doğası ve modelinizin veya algoritmanızın belirli gereksinimleri temel alınarak önemlidir. Kategorik verilerin uygun bir şekilde işlenmesi, modelinizin performansını önemli ölçüde artırabilir ve daha iyi içgörüler ve tahminler sağlayabilir.

#### Zaman Serisi Verileri(Time Series, Temporal Data)

Zaman serisi verileri, belirli bir zaman bileşeni olan bir aralık verisi biçimidir. Her gözlem veya değer, belirli bir zaman dilimi ile ilişkilidir ve bu zaman dilimi, tarihler, saatler veya her ikisinin bir kombinasyonunu içerebilir. Zaman serisi verileri, zaman içindeki trendleri, desenleri(patterns) veya ilişkileri ortaya çıkarmak için analiz edilebilir ve görselleştirilebilir.

Zamansal veri ile çalışırken dikkate alınması gereken birkaç önemli yön şunlardır:

- **Döngüsellik(Cyclicality-Periodicity)**: Zaman serisi verileri genellikle belirli zaman periyotları üzerinde tekrarlanan desenlere veya döngülere sahiptir, örneğin günlük, haftalık veya yıllık döngüler gibi. Örneğin, sıcaklıklar günlük bir döngü (sabahları serin ve öğleden sonraları sıcak) ve yıllık bir döngü (kışlar serin ve yazlar sıcak) gösterebilir.

- **Hiyerarşi(Hierarchy)**: Zaman, yıllar, çeyrekler, aylar, haftalar, günler, saatler, dakikalar ve saniyeler gibi çeşitli hiyerarşik yapılar içinde düzenlenebilir. Bu hiyerarşilerin birbirleriyle nasıl ilişkilendiğini anlamak, daha kapsamlı analiz ve karşılaştırmaları mümkün kılar.

- **Süre(Duration)**: İki zaman dilimi arasındaki fark, bir süre oluşturur ve bu da oransal verinin bir formudur. Süreler, olaylar veya dönemler arasındaki geçen zamanı daha iyi anlamak için eklenip çıkarılabilir veya karşılaştırılabilir.

#### Zamansal Veri İçin Analiz ve Görselleştirmeler

Zamansal verinin analizi, zaman içindeki trendleri ve dalgalanmaları incelemeyi içerir ve hareketli ortalamalar(moving averages), zaman tabanlı toplamlar(time-based aggregations) ve zaman serisi ayrıştırmaları(decomposition) gibi teknikleri içerebilir. İşte zamansal veri için yaygın görselleştirme teknikleri:

- **Çizgi Grafikleri (Line charts)**: Çizgi grafikleri, zamansal verileri görselleştirmek için idealdir, çünkü veri noktalarının zaman içindeki ilerlemesini gösterirler. Ardışık veri noktalarını çizgi segmentleri aracılığıyla birleştirerek, çizgi grafikleri verideki trendleri ve desenleri açıkça gösterebilir.

- **Isı Haritaları(Heatmaps)**: Isı haritaları, zamansal veriyi matris formatında görselleştirmek için kullanılabilir. Burada satırlar ve sütunlar farklı zaman birimlerini temsil eder (örneğin, saatler ve günler veya günler ve aylar), renk yoğunluğu ise belirli bir metriğin değerini gösterir. Bu görselleştirme, desenleri, trendleri ve aykırı değerleri daha yoğun bir formda ortaya koyabilir.

- **Yığılmış Alan Grafikleri(Stacked area charts)**: Yığılmış alan grafikleri, birden çok zaman serisi verisinin zaman içindeki ilerlemesini katmanlı alanlar olarak gösterir. Alanları yığıp birleştirerek, her veri serisinin kümülatif etkisini gösterir ve verinin zaman içindeki bileşimindeki trendleri ve değişiklikleri vurgular.

- **Çubuk Grafikleri(Bar charts)**: Çubuk grafikleri, çizgi grafikleri kadar yaygın olmasa da, zamansal verileri görselleştirmek için kullanılabilir. Discrete veri noktalarını veya zaman aralıkları içindeki toplam değerleri göstermek için kullanılabilir, örneğin aylık veya yıllık satışlar.

#### Zamansal Verilerde Feature Engineering

Zamansal veri genellikle makine öğrenimi modelleri ve istatistiksel analizler için değerli bilgileri etkili bir şekilde çıkarmak için özel ön işleme ve özellik mühendisliği tekniklerini gerektirir:

- **Yeniden örnekleme (Resampling)**: Zamansal veri, veri noktalarını farklı zaman aralıklarına, örneğin günlük, haftalık veya aylık olarak toplamak veya dağıtmak için yeniden örneklenmiş olabilir. Yeniden örnekleme, uzun vadeli trendleri yakalamaya veya belirli zaman aralıklarına odaklanmaya yardımcı olabilir.

- **Trend çıkarma**: Zamansal veriden trend bileşenini çıkarmak, genel değerleri düşündüğümüzde fark edilmeyen verideki desenleri ve dalgalanmaları ortaya çıkarabilir.

- **Mevsimsellik düzenleme**: Mevsimsellik genellikle zamansal veriyi günlük veya yıllık desenler gibi periyodik döngülerle etkiler. Mevsimsel etkileri düzenlemek veya kaldırmak, temel desenleri, aykırı değerleri veya ilişkileri ortaya çıkarmaya yardımcı olabilir.

- **Özellik çıkarma**: Ay, haftanın günü veya saat gibi zamanla ilgili bileşenlerden yeni özellikler oluşturmak, makine öğrenimi algoritmalarının performansını artırarak zamanla ilişkili desenler hakkında ek bilgiler yakalamaya yardımcı olabilir.

- **Eksik değerlerle başa çıkma**: Zamansal veri, lineer interpolasyon, son gözlem taşıma (Last Observation Carried Forward-LOCF) veya model tabanlı imputasyon gibi çeşitli veri tamamlama yöntemleri aracılığıyla ele alınabilen boşlukları veya eksik değerleri içerebilir.

```python
import yfinance as yf


def get_ts(sembol):
    data = yf.download(sembol, start="2017-01-01",
                        end="2017-04-30", group_by='tickers')
    data.insert(0,"id",sembol)
    data.reset_index(inplace=True)
    return data


def set_ts(semboller):
    merged_ts= pd.DataFrame()
    for sembol in semboller:
        merged_ts=pd.concat([merged_ts, get_ts(sembol)])
    return merged_ts

semboller = ["AAPL", "T", "GOOG", "MO", "AA", "AXP", "BABA","APA"]

ts_dataframes=set_ts(semboller)
ts_dataframes
```

```python
from glob import glob
csv_file_list = glob("../data/air_quality/PRSA2017/PRSA_Data_20130301-20170228/*.csv")
df = pd.DataFrame()
for file in csv_file_list:
    df = pd.concat([df,pd.read_csv(file)])
```

```python
df['timestamp'] = pd.to_datetime(df[["year", "month", "day", "hour"]])
df.drop(columns=['No'], inplace=True)
df.sort_values(by=['timestamp', 'station']).head(10)
```

```python
df2014 = df[df.timestamp.between("2014-03-01", "2014-04-01")]
df2014
df2014_limited = df2014[df2014.station.isin(['Dongsi', 'Wanliu', 'Shunyi'])]
df2014_limited
```

* `tsfresh.feature_extraction.MinimalFCParameters` basitçe hesaplanabilen birkaç özelliği içerir,
