#### Aykiri degerler ile calisma (Outliers)

Pekcok model yeni bir veri geldiginde, mevcut verilern dagilimina ait olmasini bekler. Ayni dagilima sahip olmadiginda bunu aykiri deger olarak degerlendiririz. Diger bir soylemle, aykiri deger, verinin icinde digerlerinden epeyce farkli olan degerdir. Ortalama ve varyans gibi istatistiksel parametreler aykiri degerlere karsi epeyce hassasdir. Diger taraftan, makine ogrenmesinden linear regression, adaboost gibi modellerde performansi kotu sekilde etkileyebiliyor. Dolayiysla, aykiri degerler ile ugrailmasi icin vakit haracamaya deger bir durumdur. 

Aslinda, ayriklastirma kisminda kisaca deginmistik. Ayriklastirma ile en kucuk ya da en buyuk degerleri, kullanilan ayriklastirma teknigine gore bir grup icerisinde yer alarak aykiri deger olma ozelligini kaybederek bas edebiliriz. Ya da diger taraftan, aykiri degeri bir bos deger olarak degerlendirilip onceki derslerde gordugumuz bos deger doldurma teknikleri ile halledebiliriz.

#### What we do if the data has outliers? 

- **TRIM**: Ayrkiri degerleri belirledikten sonra veri setinden basitçe kaldırmaktır.
- **WINSORIZE**: Her bir aykiri degeri en yakin normal degerlerle degistirmektir.
- **RECODE**: Aykiri degerler belirlendikten sonra onlari kaldirip, bos deger gibi dusunulup imputation islemleri yapilabillir.

#### Local Outlier Detection

Local Outlier Detection, yerel aykırı değer tespiti olarak da bilinir, veri setindeki her bir noktanın komşularını göz önünde bulundurarak bu noktanın bir aykırı değer olup olmadığını belirlemeye çalışan bir tekniktir. Bu yöntem, bir noktanın çevresindeki komşuların yoğunluğunu veya yoğunluğuna göre sapma düzeyini değerlendirerek aykırı değerleri tespit eder.

Yerel Aykırı Değer Tespiti'nin çalışma prensibi şu adımları içerir:

- **Veri Noktasının Yoğunluğu:** Bir veri noktasının komşularıyla olan ilişkisini değerlendirir. Yoğunluğu, veri noktasının komşularına olan uzaklığına dayalı olarak hesaplanır.
- **Komşuların Yoğunluğu:** Veri noktasının komşularının yoğunluğunu hesaplar. Eğer bir veri noktasının komşularının yoğunluğu düşükse, bu nokta anormal olarak kabul edilir.
- **Anormallik Puanı:** Veri noktasının anormallık puanı, komşularının yoğunluğu ile kendi yoğunluğu arasındaki farka dayalı olarak hesaplanır. Daha düşük anormallık puanları, daha büyük bir anormallik göstergesidir.

#### Isolation Forest

Isolation Forest (İzolasyon Ormanı) algoritması, anormallik tespiti için kullanılan bir başka yöntemdir. Bu algoritma, veri noktalarını rastgele bölerek “izole” eder ve bu bölünmüş yapıyı ağaç yapısıyla temsil eder. İzolasyon Ormanı, veri noktalarının normalden sapma derecesini ölçmek için kullanılır.

İşleyişini daha ayrıntılı açıklamak gerekirse:

- **Rastgele Bölme**: Bir veri noktasının özelliklerinden rastgele bir özelliği seçer ve bu özelliğin minimum ve maksimum değerleri arasında bir bölme değeri seçer.
- **Ağaç Yapısı**: Veri noktalarını bölen bu rastgele bölünmüş yapıyı ağaç olarak temsil eder. Her bölme işlemi bir düğümü ikiye böler.
- **Yol Uzunluğu**: Bir veri noktasının izole edilmesi için gereken bölme sayısı, bu ağaç yapısındaki yol uzunluğunu ifade eder. Anormal veri noktaları, daha kısa yollara sahip olma eğilimindedir.

```python
from sklearn.decomposition import PCA
from sklearn.ensemble import IsolationForest
reduction=PCA(n_components=2)
X_train_pca=reduction.fit_transform(X_train)


outlier_ISO = IsolationForest(contamination='auto')
outlier_ISO_index = outlier_ISO.fit_predict(X_train)

plt.scatter(X_train_pca.iloc[:, 0], X_train_pca.iloc[:, 1], color="k", s=3.0, label="Veri Noktaları")
plt.scatter(X_train_pca.iloc[:, 0], X_train_pca.iloc[:, 1], s=10*(outlier_ISO_index == -1), edgecolors="r", facecolors="none", label="Aykırı Değerler")
plt.xlabel(f"Aykırı Değer Sayısı: {np.sum(outlier_ISO_index == -1)}")
plt.legend()
plt.title("Isolation Forest")
plt.show()
```

#### Winsorization (Silmek yerine kurtarmak)

Winsorization veya winsorize, istatistiksel verilerdeki aşırı değerleri sınırlayarak olası yanıltıcı aykırı değerlerin etkisini azaltan bir dönüşüm yöntemidir. Bu yöntem, mühendis ve biyoistatistikçi Charles P. Winsor (1895–1951) tarafından adlandırılmıştır.

Winsorization işleyişi şu şekildedir:

- **Veri Noktalarının Yoğunluğu**: Bir veri noktasının komşularıyla olan ilişkisini değerlendirir. Yoğunluğu, veri noktasının komşularına olan uzaklığına dayalı olarak hesaplanır.
- **Komşuların Yoğunluğu**: Veri noktasının komşularının yoğunluğunu hesaplar. Eğer bir veri noktasının komşularının yoğunluğu düşükse, bu nokta anormal olarak kabul edilir.
- **Anormallik Puanı**: Veri noktasının anormallık puanı, komşularının yoğunluğu ile kendi yoğunluğu arasındaki farka dayalı olarak hesaplanır. Daha düşük anormallık puanları, daha büyük bir anormallik göstergesidir.

Winsorization, veri noktalarını anormal olarak işaretlemek için kullanılır ve genellikle aykırı değer tespiti için tercih edilir. Özellikle veri noktalarının yoğunluğuna dayalı olarak anormallık tespiti yapmak istediğinizde bu yöntem kullanışlıdır.

```python
from sklearn.linear_model import RANSACRegressor
from sklearn.linear_model import TheilSenRegressor
from sklearn.linear_model import LinearRegression


X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)

model = LinearRegression()
model.fit(X_train, y_train.to_numpy().flatten())
model.score(X_test, y_test.to_numpy().flatten())
```

```python
# outlier olanlara agirlik atamasi yapabilirz.
outlier_weights = [1 if not o else 0.05 for o in outlier_ISO.fit_predict(X_train)==-1]

# Outlier olanlara agirlik atamasi ile yonetebiliriz.
model = LinearRegression()
model.fit(X_train, y_train.to_numpy().flatten(), sample_weight=outlier_weights)
model.score(X_test, y_test.to_numpy().flatten())
```

```python
# Theil-Sen robust regression
theil = TheilSenRegressor()
model = theil.fit(X_train, y_train.to_numpy().flatten())
model.score(X_test, y_test.to_numpy().flatten())

# RANSAC robust regression
ransac = RANSACRegressor()
model = ransac.fit(X_train, y_train.to_numpy().flatten())
model.score(X_test, y_test.to_numpy().flatten())
```
