# Makine Öğrenmesi (Machine Learning) Nedir?

Makine öğrenimi, tükettikleri verilere göre öğrenen ya da performansı iyileştiren sistemler oluşturmaya odaklanan bir yapay zeka alt kümesidir.

Gorsel_1.JPG

- Veriseti : Dataset
- Örnek, Veri Noktası : Sample, Instance, Observation
- Öznitelik : Feature, Dimention
- Sınıf, Etiket : Class, Target, Label

- Kategorik

# Makineler Nasıl Öğrenir?

Makine öğreniminin altında yatan ana fikir, girdi ve (varsa) çıktı verisi kombinasyonları arasındaki mevcut matematiksel ilişkiye dayalıdır. Makine öğrenimi modeli bu ilişkiyi önceden bilmez fakat yeterli veri kümesi sağlanması halinde tahmin edebilir. Bu durum, her makine öğrenimi algoritmasının değiştirilebilir bir matematik fonksiyonu üzerine kurulduğu anlamına gelir.

# Öğrenmenin Türleri Nelerdir?

## Denetimli Öğrenme (Supervised Learning) Nedir?

Denetimli bir öğrenme algoritması bilinen bir girdi verisi seti ve verilere bilinen yanıtları alır, ardından yeni verilere yanıt için makul tahminler oluşturmak üzere bir modeli eğitir.

Gorsel_2.png

## Regresyon Nedir?

Regresyon, bağımlı bir **sayısal** değişken ile bir veya daha fazla bağımsız değişken arasındaki ilişkiyi tahmin etmek için kullanılan bir dizi istatistiksel yaklaşımdır.

Gorsel_4.png

## Sınıflandırma Nedir?

Sınıflandırma, bağımlı bir **kategorik** değişken ile bir veya daha fazla bağımsız değişken arasındaki ilişkiyi tahmin etmek için kullanılan bir dizi istatistiksel yaklaşımdır.

Gorsel_5.png

---

## Denetimsiz Öğrenme (Unsupervised Learning) Nedir?

Denetimsiz öğrenme algoritmaları, denetimli öğrenmeye kıyasla daha karmaşık işleme görevleri gerçekleştirmenizi sağlar. Dentimsiz öğrenmede sistem öğretilmiyor, verilerden öğreniyor. Denetimsiz makine öğrenimi, verilerdeki bilinmeyen her türlü paterni bulur.

Gorsel_12.jpg

## Kümeleme Nedir?

Verilerdeki doğal olarak oluşan gruplanmayı keşfetmeyi amaçlar. Supervised learning algoritmalarının aksine, yalnızca girdileri kullanarak, grup ve kümeleri tespit eder. Kümeleme algoritmaları genellikle datada bulunan benzerlikleri analiz etmek maksadıyla kullanılır.

Gorsel_6.png

## Boyut Azaltma Nedir?

Modellere girdi olarak verilen öznitelik sayısı arttıkça modelin öğrenmesi zorlaşacaktır. Var olan verisetindeki özniteliklerin sayısını belirli miktarda veri kaybı göz önüne alınarak azaltılması veya var olan öznitelikleri kullanarak yeni öznitelikler üretilerek eski özniteliklerin kullanılmaması durumları için kullanılan genel yöntemlerdir.

- Öznitelik Seçimi (Feature Selection)

Örneğin elimizde arabalar ile ilgili bir veri seti var, özniteliklerden birisi tekerlek çapı, diğeri de jant boyutu olsun. Yapılan analiz sonucunda görülmüştür ki bu ikisi birbiri ile 98% koralesyona sahip olduğu görülmüştür. Bu durumda bu iki veriden sadece birini girdi olarak kullanmak model eğitiminde bize büyük ihtimalle fayda sağlayacaktır.

Benzer bir veri seti daha düşünelim, yüzlerce öznitelik içerisinden sadece tekerlek sayısına bakmak, modelimizin çıktısının araba mı yoksa motorsiklet mi olacağını tespit etmek için yeterli olacağı için doğru özniteliğin seçimi eğitim sürecini azaltacak ve belki de başarımızı yükseltecektir.

- Öznitelik Çıkarımı (Feature Extraction)

## Pekiştirmeli/Takviyeli Öğrenme (Reinforcement Learning) Nedir?

Amaca yönelik ne yapılması gerektiğini öğrenen bir makine öğrenmesi yaklaşımıdır. Pekiştirmeli öğrenmede ajan (agent) adı verilen öğrenen makinemiz karşılaştığı durumlara bir tepki verir ve bunun karşılığında da sayısal bir ödül sinyali alır. Ajan/öğrenen makine aldığı bu ödül puanını maksimuma çıkartmak için çalışır. Bu şekilde çalışan deneme yanılma yöntemi, pekiştirmeli öğrenmenin en ayırt edici özelliğidir.

## Destek Vektör Makineleri (Support Vector Machines) Nedir?

## Temel Bileşenler Analizi (Principal Component Analysis) Nedir?

Temel amacı yüksek boyutlu verilerde en yüksek varyans ile veri setininde kullanılan öznitelikleri azaltmak ancak bunu yaparken veri kaybını minimize etmek olan **Denetimsiz Öğrenme** algortimasıdır.

PCA'in amacı özniteliklerinden oluşan her bir veri noktasını, veri setindeki tüm noktaları daha iyi/daha temiz açıklayan yeni öznitelik uzayına aktarmaktır.
