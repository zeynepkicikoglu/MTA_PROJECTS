
#### CSV (Comma Separated Values)

CSV, her veri satırının bir satır olarak gösterildiği ve veri alanlarının virgülle ayrıldığı bir düz metin dosya biçimidir. Bir CSV dosyası, tipik olarak verileri tablo veya dikdörtgen matris biçiminde içerir. İlk satır sütun adlarını veya başlıkları içerebilir. Bu basit format, CSV dosyalarının okunmasını ve yazılmasını kolaylaştırır ve uygulamalar arasında veri alışverişi için yaygın olarak kullanılırlar.

Simple CSV example:

```CSV
Name,Age,Occupation
Alice,30,Engineer
Bob,25,Scientist
Charlie,35,Artist
```

Bu örnekte, ilk satır başlıkları içerirken, takip eden her satır üç alanda (İsim, Yaş, Meslek) bir kişinin bilgilerini temsil eder. Verilerin görselleştirilmesi basittir: bunu dikdörtgen bir tablo veya elektronik tablo olarak düşünebiliriz.

##### EXCEL 

Microsoft'un EXCEL formatında ```.xlsx``` uzantılı dosyalar ile çalışmak istediğimiz pandas kütüphanesinin araçlarını kullanabiliriz. Bu örnekte, İstanbul Büyükşehirt Belediyesinin sunduğu verilerden bir örnek alalım.

````python
xls = pd.read_excel("https://data.ibb.gov.tr/dataset/321aa12d-a388-4c61-b915-522730b8230a/resource/f676e493-8995-4e50-90b3-5381da92627f/download/rayl-sistemler-hat-bazl-sefer-saylar.xlsx")
xls['Tarih'] = pd.to_datetime(xls['Tarih'])
xls.index = xls['Tarih']
xls.drop(columns="Tarih", inplace=True)
xls.fillna(0, inplace=True)
metrolar = ["M"+str(i)for i in range(1,6)]
kismi_xls=xls.loc[:, metrolar]
kismi_xls
````

````python
%matplotlib inline
kismi_xls.plot(figsize=(20,4))
````

##### SQL, JSON, PARQUET

```sqlite3 sqlalchemy duckdb mongodb``` gibi pythonda ilişkisel olan ve olmaayan veri tabanları için paketler mevcut. Veri yapıları konusunda detaylı şekilde ele anılanacaktır. Bizler burada küçük bir gösterim yapacağız.


````python

import os
from sqlalchemy import create_engine

DataFolderName="../data/data-type"

def rfd(filename):
    return os.path.join(DataFolderName, filename)

if not os.path.exists(DataFolderName):
    os.mkdir(DataFolderName)

os.listdir(DataFolderName)
````

##### Elimizde ki veriyi farkli formatlari da kayit edelim

````python
kismi_xls.to_csv(rfd("csv-example.csv"))
kismi_xls.to_json(rfd("json-example.json"))
kismi_xls.to_parquet(rfd("parquet-example.parquet"), engine='auto',index=False)
````

````python
# Bir swl database olusturalim ve icerisi dataframe'leri gonderelim
engine=create_engine(os.path.join("sqlite:///", DataFolderName, "sql-example.sqlite"), echo=True)
kismi_xls.to_sql("kismi-veri", con=engine, index=False)

# birden fazla dataframe ekleyeabiliriz:
engine=create_engine(os.path.join("sqlite:///", DataFolderName, "sql-example-multi.sqlite"), echo=True)
kismi_xls.to_sql("kismi-veri", con=engine, index=False)
xls.to_sql("full-veri", con=engine, index=False)
````

````python
engine=create_engine(os.path.join("sqlite:///", DataFolderName, "sql-example.sqlite"), echo=False)
data_sql = pd.read_sql("kismi-veri", con=engine)
data_sql.head()
````

#### JSON

JSON, insan tarafından okunabilir olması için tasarlanmış metin tabanlı bir formattır. JavaScript'ten türetilmiştir, ancak bağımsız olarak ve birçok programlama dili arasında bir veri değişim formatı olarak kullanılabilir. JSON, verileri depolamak için anahtar-değer çiftleri kullanır ve doğası gereği hiyerarşiktir, yani veriler görsel olarak bir ağaç yapısı olarak temsil edilebilir.

```
[
  {
    "Name": "Alice",
    "Age": 30,
    "Occupation": "Engineer"
  },
  {
    "Name": "Bob",
    "Age": 25,
    "Occupation": "Scientist"
  },
  {
    "Name": "Charlie",
    "Age": 35,
    "Occupation": "Artist"
  }
]
```
Bu örnekte, her bir kişinin bilgileri anahtar-değer çiftleri içeren bir nesne olarak saklanır ve tüm veri kümesi bir JSON dizisi olarak temsil edilir. CSV'den farklı olarak, JSON verileri dikdörtgen bir tablo yerine bir ağaç yapısı olarak görselleştirilir.

#### PARQUET
Parquet veri tipi, karmaşık iç içe veri yapılarına sahip iken temelde csv formatına dayanmaktadır. Ancak, diskte yer alması konusunda parquet daha az boyuta sahip olması ile başarılıdır. Diğer taraftan, csv dosyası basitçe açılıp değişiklik yapılabilinirken parquet karmaşık veri yappılarını tutmak için sütun bazlı verileri saklar ve csv kadar kolay değlidir. 

Veri biliminde verilerinizi bir veri tabanında ya da bir bulut sağlayıcısında tutmanız gerektiğinde, CSV dosya formatı epey pahalıya gelecektir.

#### XML

XML, HTML'ye benzer şekilde, verilerin yapısını tanımlamak için etiketler kullanan bir işaretleme dilidir. XML hiyerarşiktir ve bir XML belgesinde yer alan veriler görsel olarak bir ağaç yapısı şeklinde gösterilebilir. XML çok yönlüdür ve çok çeşitli veri yapılarını temsil etmek için kullanılabilir.

```xml
<people>
  <person>
    <Name>Alice</Name>
    <Age>30</Age>
    <Occupation>Engineer</Occupation>
  </person>
  <person>
    <Name>Bob</Name>
    <Age>25</Age>
    <Occupation>Scientist</Occupation>
  </person>
  <person>
    <Name>Charlie</Name>
    <Age>35</Age>
    <Occupation>Artist</Occupation>
  </person>
</people>
```

XML'in birçok çeşidi vardır. XML'in en ünlü ve yaygın olarak kullanılan çeşidi [Hypertext Markup Language (HTML)](https://en.wikipedia.org/wiki/HTML)'dir. XML başlıca bir ders olarak düşünebiliriz. Küçük bir örnek göstereceğiz. Python ile XML ile uğraşmak için en güvenli yöntem [xmltodict](https://github.com/martinblech/xmltodict) kütüphanesi aracılığıyla onu bir sözlüğe (veya bir hash haritasına) dönüştürmektir.


````python
with urlopen("http://www.ecb.europa.eu/stats/eurofxref/eurofxref-daily.xml") as url:
    raw = url.read()
````

## YAML (YAML Ain't Markup Language)

YAML insan tarafından okunabilir bir veri serileştirme formatıdır. Verilerin hiyerarşik yapısını temsil etmek için girinti ve basit noktalama işaretleri kullanır. YAML'nin okunması ve yazılması kolaydır ve genellikle yapılandırma dosyaları, diller arasında veri alışverişi ve veri depolama için kullanılır.

```yaml
- Name: Alice
  Age: 30
  Occupation: Engineer
- Name: Bob
  Age: 25
  Occupation: Scientist
- Name: Charlie
  Age: 35
  Occupation: Artist
```
Bu örnekte, her öğe bir kısa çizgi ve girinti ile temsil edilmektedir. JSON ve XML gibi, YAML verileri de bir ağaç yapısı olarak görselleştirilir.

YAML genellikle config dosyaları için kullanılır.

### Yapısız veri formatları
Yapılandırılmamış veri dosyaları bir dizi bitten oluşur. Dosya, bitleri birbirinden herhangi bir şekilde ayırmaz. Dosyanın içine bakıp herhangi bir yapı göremezsiniz çünkü görecek bir şey yoktur. Yapılandırılmamış dosya formatları dosya kullanıcısının verileri nasıl yorumlayacağını bilmesine bağlıdır. Belirli bir düzen veya şemaya sahip olmayan(tabular olmayan), çeşitli türde verileri ifade etmek için kullanılan format [yapısız veri](https://en.wikipedia.org/wiki/Unstructured_data) olarak tanımlanabilir. Örneğin resim, video, ses kayıtları, sensör kayıtları, metin vb olabilmektedir.


Python'da [unstructured](https://unstructured-io.github.io/unstructured/index.html) kütüphanesi yapısız veri türleri için ön hazırlık aşamsında kullanılmak üzere oluşturulmuş paketlerden bir tanesidir.

````python
#https://nipy.org/nibabel/coordinate_systems.html
def show_slices(slices):
   """ Function to display row of image slices """
   fig, axes = plt.subplots(1, len(slices))
   for i, slice in enumerate(slices):
       axes[i].imshow(slice.T, cmap="gray", origin="lower")

slice_0 = mri_veri[26, :, :]
slice_1 = mri_veri[:, 30, :]
slice_2 = mri_veri[:, :, 16]
show_slices([slice_0, slice_1, slice_2])
plt.suptitle("Axis0, Axis1, Axis 2")  
````
