#### Kategorik degiskenlerde esik degeri belirleme

#### Dogrusal Iliskiler

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
x,y = make_regression(n_samples=1000,n_features=1, n_targets=1, noise=5.5)
#y = y[:, np.newaxis]
data=pd.DataFrame([x.flatten(), y]).T
data.columns=['x', 'y']
data.head()
```

```python
import seaborn as sns
sns.lmplot(x='x', y='y', data=data, order=1)
plt.ylabel('Target')
plt.xlabel('Independent variable')
```

```python
# scikit-learn verileri dataframe formatinda kabul eder.
model=LinearRegression()
model.fit(data[['x']], data[['y']])
```

```python
tahminler = model.predict(data[['x']])
hatalar = data[['y']]-tahminler
```

```python
fig=plt.figure()
plt.scatter(x=data[['x']], y=hatalar)
plt.axhline(y=0, color='red')
plt.ylabel('hatalar')
plt.xlabel('x degiskeni')
```

```python
sns.displot(hatalar,kind='hist', kde=True)
```

```python
import scipy.stats as stats
_,_= stats.probplot(data['x'], dist="norm", plot=plt)
```

#### Dagilimlari Ogrenme ve Aykiri Degerleri Bulma

```python
from distfit import distfit
dfit = distfit()
results = dfit.fit_transform(hatalar.to_numpy())
results['summary']
dfit.plot(fontsize=20, figsize=(12, 5))
```

#### Aykiri Degerler (Outliers)

```python
from sklearn.datasets import load_wine
data = load_wine(as_frame=True)
features=data['data'][['alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash']]
target=data['target']
data=pd.concat([features,target], axis=1)
data.head()
```

```python
sns.pairplot(data, hue='target',kind='scatter', diag_kind='kde')
```

```python
sns.violinplot(data=data, scale="width")
```

```python
sns.boxplot(data=data)
```

```python
def find_boundaries(df, variable, distance):
    IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)
    lower_boundary = df[variable].quantile(0.25) - (IQR * distance)
    upper_boundary = df[variable].quantile(0.75) + (IQR * distance)
    return upper_boundary, lower_boundary
```

```python
variable='alcalinity_of_ash'
upper_boundary, lower_boundary = find_boundaries(data, variable, 1.5)

outliers_index = (data[variable]>upper_boundary) | (data[variable]<lower_boundary) 
data[outliers_index]
```

```python
data[~outliers_index].reset_index()
```
